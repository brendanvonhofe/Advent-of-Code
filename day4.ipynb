# Part 2

```python
# byr (Birth Year) - four digits; at least 1920 and at most 2002.
# iyr (Issue Year) - four digits; at least 2010 and at most 2020.
# eyr (Expiration Year) - four digits; at least 2020 and at most 2030.
# hgt (Height) - a number followed by either cm or in:
# If cm, the number must be at least 150 and at most 193.
# If in, the number must be at least 59 and at most 76.
# hcl (Hair Color) - a # followed by exactly six characters 0-9 or a-f.
# ecl (Eye Color) - exactly one of: amb blu brn gry grn hzl oth.
# pid (Passport ID) - a nine-digit number, including leading zeroes.
# cid (Country ID) - ignored, missing or not.
```

```python
def byr(s):
    if not re.compile('\d{4}').fullmatch(s):
        return False
    if int(s) < 1920 or int(s) > 2002:
        return False
    return True

def iyr(s):
    if not re.compile('\d{4}').fullmatch(s):
        return False
    if int(s) < 2010 or int(s) > 2020:
        return False
    return True

def eyr(s):
    if not re.compile('\d{4}').fullmatch(s):
        return False
    if int(s) < 2020 or int(s) > 2030:
        return False
    return True

def hgt(s):
    if re.compile('\d+(cm|in)').fullmatch(s):
        if re.compile('\d+cm').fullmatch(s):
            if int(s[:-2]) >= 150 and int(s[:-2]) <= 193:
                return True
            else:
                return False
        else:
            if int(s[:-2]) >= 59 and int(s[:-2]) <= 76:
                return True
            else:
                return False
    else:
        return False
    
def hcl(s):
    if re.compile('#(\d|[a-f]){6}').fullmatch(s):
        return True
    else:
        return False

def ecl(s):
    if s in ['amb', 'blu', 'brn', 'gry', 'grn', 'hzl', 'oth']:
        return True
    else:
        return False
    
def pid(s):
    if re.compile('\d{9}').fullmatch(s):
        return True
    else:
        return False
```

```python
required_passport_field_checks = {
    'byr': byr,
    'iyr': iyr,
    'eyr': eyr,
    'hgt': hgt,
    'hcl': hcl,
    'ecl': ecl,
    'pid': pid
}
```

```python
n_valid_passports = 0
for passport in puzzle_input:
    fields = []
    for chunk in passport.split('\n'):
        fields += chunk.split(' ')
        
    fields = {field.split(':')[0]: field.split(':')[1] for field in fields}
    
    valid = 1
    for required_field in required_passport_fields:
        if required_field not in fields:
            valid = 0
        else:
            if not required_passport_field_checks[required_field](fields[required_field]):
                valid = 0
            
    if valid:
        n_valid_passports += 1
        
print(n_valid_passports)
```
