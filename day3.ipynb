```python
# Right 1, down 1.
# Right 3, down 1. (This is the slope you already checked.)
# Right 5, down 1.
# Right 7, down 1.
# Right 1, down 2.
```

```python
from functools import reduce
```

```python
tree = '#'
slopes = [
    (1, 1),
    (3, 1),
    (5, 1),
    (7, 1),
    (1, 2)
]
```

```python
n_trees_encountered_per_slope = []
for right, down in slopes:
    n_trees_encountered = 0
    x_pos = y_pos = 0
    
    while y_pos < len(puzzle_input):
        row = puzzle_input[y_pos]
        while x_pos >= len(row):
            row *= 2
        if row[x_pos] == tree:
            n_trees_encountered += 1
        x_pos += right
        y_pos += down

    n_trees_encountered_per_slope.append(n_trees_encountered)
    
print(reduce(lambda x, y: x * y, n_trees_encountered_per_slope))
```
